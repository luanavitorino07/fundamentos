```python
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('portuguese'))


palavras_remover = {"sobre", "São", "todos", "Paulo", "Rio", "R"}

stop_words.update(palavras_remover)

df_infos
titulos = ' '.join(df_infos['titulo'])
titulos_sem_pontuacao = re.sub(r'[^\w\s]', '', titulos)
palavras = titulos_sem_pontuacao.split()


palavras_filtradas = [palavra for palavra in palavras if palavra not in stop_words]

frequencia_palavras = Counter(palavras_filtradas)

mais_comuns = frequencia_palavras.most_common(25)
mais_comuns
```

```python
df_mais_comuns = pd.DataFrame(mais_comuns, columns=['palavra', 'frequencia'])
df_mais_comuns
grafico_mais_comuns = px.bar(df_mais_comuns, x='palavra', y='frequencia')
grafico_mais_comuns
```

##Citações a presidentes do Brasil nos parágrafos

```python
presidentes = ['Dilma','Temer','Bolsonaro', 'Lula', 'Deus', 'FHC', 'Collor', 'Itamar']


def contar_presidentes(texto, lista_termos):
    paragrafos = ' '.join(texto)

    contador = Counter()
    for palavra in presidentes:
        contador[palavra] = paragrafos.lower().count(palavra.lower())
    return contador

# Aplicar a função à coluna 'paragrafos' e somar as ocorrências
frequencias_totais = Counter()
for paragrafo in df_infos['paragrafos']:
    frequencias_totais += contar_presidentes(paragrafo, presidentes)


frequencias_totais.values()

grafico_1 = px.bar(x = frequencias_totais.keys(), y=frequencias_totais.values())
grafico_1
```

## Citações a anos nos paragrafos

```python
anos = ['2014','2015','2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023','2024']


def contar_anos(texto, anos):

    paragrafos = ' '.join(texto)

    contador = Counter()
    for ano in anos:
        contador[ano] = paragrafos.count(ano)
    return contador

frequencias_totais = Counter()
for paragrafo in df_infos['paragrafos']:
    frequencias_totais += contar_anos(paragrafo, anos)


frequencias_totais
```
